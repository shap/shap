# `TextGeneration` model wrapper

Wrapping your model with `shap.models.TextGeneration` makes it ready for SHAP explanations. It handles the "heavy lifting" of generating tokens and caching outputs so that the explanation process runs smoothly.

This notebook shows you how to set it up in just a few steps.

```python
import shap
import transformers
import numpy as np

# First, let's load a small model and its tokenizer.
# We're using a 'tiny' version here to keep things fast!
model_name = "sshleifer/tiny-gpt2"
tokenizer = transformers.AutoTokenizer.from_pretrained(model_name)
model = transformers.AutoModelForCausalLM.from_pretrained(model_name)

# If you have a GPU, you can set device="cuda" here.
device = "cpu"
```

## Setting up the wrapper

Initializing the wrapper is as simple as passing in your model and tokenizer. If it's a Hugging Face model, SHAP knows exactly how to work with it.

```python
text_gen_model = shap.models.TextGeneration(model, tokenizer, device=device)

X = "The capital of France is"

# When we call the model, it generates the next tokens for us.
target_tokens = text_gen_model(X)

print(f"Input text: {X}")
print(f"Generated token IDs: {target_tokens}")
print(f"Decoded output: {tokenizer.decode(target_tokens[0])}")
```

## How it helps with explanations

Usually, SHAP handles this wrapper for you behind the scenes. But it's good to know that it keeps track of things like the "explanation row" to make sure every sentence in your dataset gets its own fair share of processing.

```python
# The wrapper keeps a counter for the rows it has processed.
print(f"Current explanation row: {text_gen_model.explanation_row}")

# Every time we give it a new sentence, the counter goes up!
text_gen_model("Let's try one more example.")
print(f"Row counter after the next call: {text_gen_model.explanation_row}")
```
