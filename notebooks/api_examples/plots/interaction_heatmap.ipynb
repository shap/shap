# `interaction_heatmap` plot

This notebook demonstrates how to use `shap.plots.interaction_heatmap` to visualize pairwise interaction strength between features.

```python
import xgboost

import shap

# Train a simple model on the Adult dataset
X, y = shap.datasets.adult()
X_small = X.iloc[:600]
y_small = y[:600]

model = xgboost.XGBClassifier(
    n_estimators=80, max_depth=3, learning_rate=0.08, subsample=0.9, colsample_bytree=0.9, random_state=0
).fit(X_small, y_small)

explainer = shap.Explainer(model, X_small)
shap_values = explainer(X_small.iloc[:300])
```

The heatmap below uses a heuristic interaction score based on how feature values track SHAP value variation in binned regions.

- Darker colors indicate stronger estimated interaction strength.
- Diagonal entries are zero by construction.

```python
shap.plots.interaction_heatmap(shap_values, max_display=10)
```

You can also retrieve ranked interaction pairs programmatically with `shap.utils.rank_interactions`:

```python
pairs = shap.utils.rank_interactions(shap_values, max_pairs=10)
feature_names = list(shap_values.feature_names)

for i, j, score in pairs[:5]:
    print(f"{feature_names[i]} x {feature_names[j]}: {score:.3f}")
```
