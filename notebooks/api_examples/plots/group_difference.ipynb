# group_difference plot

Demonstrates `shap.plots.group_difference` with a synthetic `shap.Explanation`.


```python
import numpy as np

import shap

rng = np.random.default_rng(0)
values = rng.normal(size=(300, 6))
feature_names = [f"Feature {i}" for i in range(values.shape[1])]
shap_values = shap.Explanation(values=values, feature_names=feature_names)
group_mask = rng.random(size=values.shape[0]) > 0.5
```

```python
shap.plots.group_difference(shap_values, group_mask, max_display=6)
```

```python
# Compare output shift between groups (1D Explanation)
shap.plots.group_difference(shap_values.sum(1), group_mask, xlabel="Group output shift")
```
