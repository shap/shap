# embedding plot

Demonstrates `shap.plots.embedding` with a synthetic `shap.Explanation`.


```python
import numpy as np

import shap

rng = np.random.default_rng(0)
values = rng.normal(size=(200, 5))
feature_names = [f"Feature {i}" for i in range(values.shape[1])]
shap_values = shap.Explanation(values=values, feature_names=feature_names)
```

```python
# Default: PCA to 2D, color by feature 0
shap.plots.embedding(0, shap_values)
```

```python
# Color by the per-row sum of SHAP values
shap.plots.embedding("sum()", shap_values, alpha=0.8)
```
